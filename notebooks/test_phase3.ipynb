```python
import sys
sys.path.append('/teamspace/studios/this_studio/financial_advisory_system')

from src.analytics.feature_engineering import FeatureEngineer

# Initialize and test
engineer = FeatureEngineer()

# Test with same stocks as previous components
test_symbols = ['MSFT', 'GOOG', 'TSLA', 'NVDA', 'ABBV', 'CVX']

print("⚙️ PHASE 3: Feature Engineering Pipeline")
print("=" * 50)
print(f"🔧 Creating ML features for: {', '.join(test_symbols)}")

results = {}
for symbol in test_symbols:
    print(f"\n🔍 Processing {symbol}...")
    success = engineer.process_stock_features(symbol)
    results[symbol] = success
    
    if success:
        print(f"   ✅ {symbol}: SUCCESS")
    else:
        print(f"   ❌ {symbol}: FAILED")

# Summary
successful = sum(results.values())
print(f"\n📈 RESULTS SUMMARY:")
print(f"   Processed: {len(test_symbols)} stocks")
print(f"   Successful: {successful}")
print(f"   Failed: {len(test_symbols) - successful}")

if successful > 0:
    print(f"\n✅ Feature engineering pipeline working!")
    print(f"📊 ML features stored in 'ml_features' table")
    print(f"🎯 PHASE 3 COMPLETE!")
```
